```
#1. 패키지 가져오기
import numpy as np

print(np.__version__)
```

```
#2. 데이터 만들기

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[0],[0],[1]])

print(X)
print(y)
```

```
#3. Keras 패키지 불러오기

from tensorflow import keras
from tensorflow.keras import layers

print(keras.__version__)
```

```
#4. Network 설정

model = keras.Sequential() # 빈 모델

model.add(layers.Dense(units= 1,input_dim=2)) #단일 퍼셉트론(입력받자마자 출력)
model.add(layers.Activation('sigmoid'))

model.summary() #parameter = weight + bias (input*units + 1)
```

```
#5. compile(Loss+Optimizer)

model.compile(loss='binary_crossentropy',
              optimizer='sgd')
```

```
#6.모델 학습

model.fit(X,y,batch_size= 1,epochs=200)
```

```
#7. 모델 Test

test_data = np.array([[0,1],[1,1],[0,0]])

result = model.predict(test_data)
print(result)
```

```
result2 = model.predict_classes(test_data)
print(result2)
```

```
#8. 모델 살펴보기

print(model.get_weights())
```
