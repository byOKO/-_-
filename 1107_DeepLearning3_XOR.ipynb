** 11월 7일 XOR Function - MLP **

```
#1. numpy 가져오기

import numpy as np

print(np.__version__)
```

```
#2. 데이터 만들기

X = np.array([[0,0],[1,0],[0,1],[1,1]])
y = np.array([[0],[1],[1],[0]])

print(X.shape)
print(y.shape)
```

```
#3. keras 가져오기

from tensorflow import keras

print(keras.__version__)
```

```
#4. Network 구성하기

model = keras.Sequential()

model.add(keras.layers.Dense(8, input_dim=2, activation='relu'))
model.add(keras.layers.Dense(1,activation='sigmoid'))

model.summary()
```

```
#5. compile

model.compile(loss='binary_crossentropy',
              optimizer='sgd')
```

```
#6. 모델학습
model.fit(X,y,batch_size=1,epochs=700)
```

```
#7. Test

test_data = np.array([[0,0],[1,1],[0,1],[1,0]])
print(model.predict(test_data))
```

```
result = model.predict_classes(test_data)
print(result)
```
